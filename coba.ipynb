```python
import folium

# Koordinat lokasi awal (misalnya: Jakarta)
start_location = [-6.2088, 106.8456]

# Koordinat lokasi tujuan (misalnya: Bandung)
destination_location = [-6.9175, 107.6191]

# Membuat peta
map = folium.Map(location=start_location, zoom_start=7)

# Menambahkan marker untuk lokasi awal
folium.Marker(
    location=start_location,
    popup='Lokasi Awal: Jakarta',
    icon=folium.Icon(color='blue')
).add_to(map)

# Menambahkan marker untuk lokasi tujuan
folium.Marker(
    location=destination_location,
    popup='Lokasi Tujuan: Bandung',
    icon=folium.Icon(color='red')
).add_to(map)

# Menggambar garis antara lokasi awal dan tujuan
folium.PolyLine(
    locations=[start_location, destination_location],
    color='green',
    weight=5,
    opacity=0.7
).add_to(map)

# Menampilkan peta
map
```

```python
import heapq

def dijkstra(graph, start, target):
    # Inisialisasi jarak awal dan priority queue
    distances = {node: float('inf') for node in graph}
    distances[start] = 0
    priority_queue = [(0, start)]  # (jarak, node)
    predecessors = {node: None for node in graph}

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        # Jika node sudah diproses, lewati
        if current_distance > distances[current_node]:
            continue

        # Perbarui jarak ke tetangga
        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight

            # Jika jarak lebih kecil, perbarui
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                predecessors[neighbor] = current_node
                heapq.heappush(priority_queue, (distance, neighbor))

    # Bangun jalur dari predecessors
    path = []
    current = target
    while current is not None:
        path.append(current)
        current = predecessors[current]
    path.reverse()

    return path, distances[target]

# Representasi graf dari data
graph = {
    "Klinik Unesa": {"A1": 16},
    "A1": {"A2": 35, "B1": 69},
    "A2": {"A3": 25},
    "A3": {"A4": 10},
    "A4": {"A5": 40},
    "A5": {"A6": 101},
    "A6": {"A7": 520},
    "A7": {"A8": 140},
    "A8": {"RS BUNDA": 525},
    "B1": {"B2": 58.4},
    "B2": {"B3": 17},
    "B3": {"B4": 23},
    "B4": {"B5": 95},
    "B5": {"B6": 360},
    "B6": {"B7": 15},
    "B7": {"B8": 127},
    "B8": {"B9": 108},
    "B9": {"B10": 92.5},
    "B10": {"B11": 47},
    "B11": {"A8": 28}
}

# Menjalankan algoritma Dijkstra
start = "Klinik Unesa"
target = "RS BUNDA"
path, total_distance = dijkstra(graph, start, target)

# Output hasil
print("Jalur terpendek:", " -> ".join(path))
print("Total jarak:", total_distance)
```
